Ejercicio

Usando Python, encuentra el exponente
 más pequeño de tal manera que el número complejo (3+4i)^n
 tenga módulo mayor que 100.

```python
abs(3+4j)
```

```python
z = 3 + 4j
```

```python
m=100
```

```python
def find_n(z,m):
    for n in range(1, m+1):
          if abs(z**n)>m:
            return n
    raise ValueError(f'no hay exponente n tal que {z}^n sea mayor que {m}' )
```

```python
n=find_n(z,m)
print(f'el exponente n tal que {z}^n es mayor que {m} es {n}' )
```
